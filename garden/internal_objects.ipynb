## Plant Library Objects\nBuild the plant taxonomy: PlantType → PlantVariety → companions, pests, diseases

```python
# --- Plant Types ---
tomato_type = PlantType(
    name="Tomato",
    category=garden_enum.PlantCategory.VEGETABLE,
)

basil_type = PlantType(
    name="Basil",
    category=garden_enum.PlantCategory.HERB,
)

fennel_type = PlantType(
    name="Fennel",
    category=garden_enum.PlantCategory.HERB,
)

print(tomato_type)
print(basil_type)
print(fennel_type)
```

```python
# --- Plant Variety ---
roma_tomato = PlantVariety(
    plant_type_id=tomato_type.plant_type_id,
    variety_name="Roma",
    plant_category=garden_enum.PlantCategory.VEGETABLE,
    sun_tolerance=garden_enum.SunTolerance.FULL_SUN,
    water_required=garden_enum.WaterRequired.MEDIUM,
    soil_n=0.5,
    soil_p=0.3,
    soil_k=0.4,
    growth_needs="Deep watering twice weekly, stake when 12 inches tall",
    post_harvest_soil_needs="Add compost, rest bed or follow with legumes",
    days_to_harvest=75,
    indoor_sow_weeks_before_frost=6,
    outdoor_sow_date_range="May 15 - Jun 1",
    spacing_inches=24.0,
    harvest_timing="When deep red and slightly soft to touch",
    temp_min_air_f=50.0,
    temp_min_ground_f=60.0,
    height_inches_estimate=48.0,
)

print(roma_tomato)
```

```python
# --- Companions & Antagonists ---
tomato_companions = PlantCompanions(
    plant_type_id=tomato_type.plant_type_id,
    companions=[basil_type.plant_type_id],
    antagonists=[fennel_type.plant_type_id],
)

print(f"Tomato companions:  {tomato_companions.companions}")
print(f"Tomato antagonists: {tomato_companions.antagonists}")

# --- Pests & Diseases ---
tomato_pest = PlantPest(
    plant_type_id=tomato_type.plant_type_id,
    pest_name="Tomato Hornworm",
    symptoms="Large holes in leaves, defoliation, dark droppings on leaves",
    treatment="Hand pick, introduce parasitic wasps, apply Bt if severe",
)

tomato_disease = PlantDisease(
    plant_type_id=tomato_type.plant_type_id,
    disease_name="Early Blight",
    symptoms="Dark concentric rings on lower leaves, yellowing around spots",
    treatment="Remove affected leaves, apply copper fungicide, improve air circulation",
)

print(tomato_pest)
print(tomato_disease)
```

## Garden State Objects\nBuild a Garden → Plot → PlantInstance hierarchy

```python
from datetime import date

# --- Garden ---
my_garden = Garden(
    name="Backyard Garden",
    zipcode="07960",
    aerial_image_path="data/images/aerial.jpg",
    lat=40.7968,
    lon=-74.4815,
    created_at=date.today(),
)

print(my_garden)
```

```python
# --- Plot with sun and height zone regions ---
raised_bed = Plot(
    garden_id=my_garden.garden_id,
    name="South Raised Bed",
    polygon=[(100, 200), (300, 200), (300, 400), (100, 400)],
    area_sqft=32.0,
    sun_zone_default=garden_enum.SunZone.FULL_SUN,
    height_zone_default=garden_enum.HeightZone.MEDIUM,
    sun_zone_regions=[
        SunZoneRegion(
            sun_zone=garden_enum.SunZone.FULL_SUN,
            polygon=[(100, 200), (200, 200), (200, 400), (100, 400)],  # left half
        ),
        SunZoneRegion(
            sun_zone=garden_enum.SunZone.PARTIAL_SHADE,
            polygon=[(200, 200), (300, 200), (300, 400), (200, 400)],  # right half
        ),
    ],
    height_zone_regions=[
        HeightZoneRegion(
            height_zone=garden_enum.HeightZone.TALL,
            polygon=[(100, 200), (300, 200), (300, 280), (100, 280)],  # back row
        ),
        HeightZoneRegion(
            height_zone=garden_enum.HeightZone.LOW,
            polygon=[(100, 280), (300, 280), (300, 400), (100, 400)],  # front row
        ),
    ],
    notes="Best sun exposure on left half, shaded by fence on right",
)

my_garden.plots.append(raised_bed)

print(f"Plot:           {raised_bed.name}")
print(f"Area:           {raised_bed.area_sqft} sqft")
print(f"Sun default:    {raised_bed.sun_zone_default.value}")
print(f"Height default: {raised_bed.height_zone_default.value}")
print(f"Sun regions:    {[r.sun_zone.value for r in raised_bed.sun_zone_regions]}")
print(f"Height regions: {[r.height_zone.value for r in raised_bed.height_zone_regions]}")
print(f"Plant instances: {raised_bed.plant_instances}  ← empty until next cell")
print(f"\nGarden now has {len(my_garden.plots)} plot(s)")
```

```python
# --- Plant Instances — three lifecycle statuses ---

# PLANNED: decided to plant, sow date set, not yet in ground
tomato_1 = PlantInstance(
    plot_id=raised_bed.plot_id,
    plant_id=roma_tomato.plant_id,
    plant_type_id=tomato_type.plant_type_id,
    location_x=120.0,
    location_y=220.0,
    status=garden_enum.PlantStatus.PLANNED,
    planned_sow_date=date(2024, 5, 28),
)

# ACTIVE: in the ground
tomato_2 = PlantInstance(
    plot_id=raised_bed.plot_id,
    plant_id=roma_tomato.plant_id,
    plant_type_id=tomato_type.plant_type_id,
    location_x=180.0,
    location_y=220.0,
    status=garden_enum.PlantStatus.ACTIVE,
    planned_sow_date=date(2024, 5, 20),
    planted_date=date(2024, 5, 22),
    harvest_count=12,
    harvest_weight_lbs=4.5,
)

# REMOVED: pulled out
tomato_3 = PlantInstance(
    plot_id=raised_bed.plot_id,
    plant_id=roma_tomato.plant_id,
    plant_type_id=tomato_type.plant_type_id,
    location_x=240.0,
    location_y=220.0,
    status=garden_enum.PlantStatus.REMOVED,
    planned_sow_date=date(2024, 5, 20),
    planted_date=date(2024, 5, 22),
    removed_date=date(2024, 9, 1),
)

raised_bed.plant_instances.extend([tomato_1, tomato_2, tomato_3])

# traverse the full tree
for plot in my_garden.plots:
    print(f"Plot: {plot.name} — {len(plot.plant_instances)} plant(s)")
    for p in plot.plant_instances:
        print(f"  {p.instance_id[:8]}... | status: {p.status.value:8} | active: {p.is_active} | harvested: {p.harvest_weight_lbs} lbs  | planned_sow: {p.planned_sow_date}")
```

## Photos & Snapshots

```python
import json
from dataclasses import asdict

# --- Photo ---
photo = PlantPhoto(
    plant_instance_id=tomato_1.instance_id,
    plant_type_id=tomato_type.plant_type_id,
    plant_id=roma_tomato.plant_id,
    plant_type="tomato",
    name_of_file=f"{tomato_1.instance_id}_2024-06-15.jpg",
    date_added=date(2024, 6, 15),
    notes="First flowers appearing",
)

print(f"Stored at: data/images/{photo.plant_type}/{photo.name_of_file}")

# --- Snapshot ---
# asdict(my_garden) captures the full tree: garden → plots → plant_instances
snapshot = GardenSnapshot(
    snapshot_date=date.today(),
    triggered_by=garden_enum.SnapshotTrigger.MANUAL,
    snapshot_json=json.dumps(asdict(my_garden), default=str),
    is_current=True,
)

print(f"\nSnapshot id:      {snapshot.snapshot_id[:8]}...")
print(f"Snapshot date:    {snapshot.snapshot_date}")
print(f"Triggered by:     {snapshot.triggered_by.value}")
print(f"Is current:       {snapshot.is_current}")
print(f"JSON size:        {len(snapshot.snapshot_json)} chars")

# verify round-trip — confirm plant instances are nested inside plots
parsed = json.loads(snapshot.snapshot_json)
print(f"\nPlots in snapshot:           {len(parsed['plots'])}")
print(f"Plant instances in plot[0]:  {len(parsed['plots'][0]['plant_instances'])}")
```
